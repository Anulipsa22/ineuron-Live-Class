```python
#Merge Operation
data4 = {'emp_id': [101,102,103,104],
         'salary' : [234,345,200,400],
         'pf' : [23,34,56,78] 
         }

data5 = {'emp_id' : [101,102,104,105],
         'mob_no' : [3434,5656,767,878],
         'house_no' : [345,345,345,56]
         }

df4 = pd.DataFrame(data4)
df5 = pd.DataFrame(data5)
```

```python
pd.merge(df4,df5) #Basically performs inner join on Common Column (Emp_id)
```

```python
pd.merge(df4,df5,how='left') #Left Join
```

```python
pd.merge(df4,df5,how='right') #Right Join
```

```python
pd.merge(df4,df5,how='right',on='emp_id')
```

```python
data6 = {'emp_id1': [101,102,103,104],
         'salary' : [234,345,200,400],
         'pf' : [23,34,56,78] 
         }

data7 = {'emp_id2' : [101,102,104,105],
         'mob_no' : [3434,5656,767,878],
         'house_no' : [345,345,345,56]
         }

df6 = pd.DataFrame(data6)
df7 = pd.DataFrame(data7)

pd.merge(df6,df7,left_on='emp_id1',right_on='emp_id2')
```

```python
data8 = {'emp_id' : [101,102,103,104],
         'salary' : [234, 345,200,400],
         'pf' : [23,34,56,78],
         }
         
data9  = {'emp_id' : [101,102,104,105],
          'salary' : [234, 345,200,878],
          'house_nu' : [345,345,345,56]
          }

df8 = pd.DataFrame(data8)
df9 = pd.DataFrame(data9)

pd.merge(df8,df9,on=['emp_id','salary']) #Joining on Multiple Columns
```

```python
#Join Operations
data10 = {'emp_id1' : [101,102,103,104],
         'salary1' : [234, 345,200,400],
         'pf' : [23,34,56,78],
         }
         
data11  = {'emp_id2' : [101,102,104,105,123],
          'salary2' : [234, 345,200,878,43],
          'house_nu' : [345,345,345,56,234]
          }

df10 = pd.DataFrame(data10,index=['a','b','c','d'])
df11 = pd.DataFrame(data11,index=['a','b','d','e','f'])

df10.join(df11) #Match basis index. It will perform join basis row index whereas in Merge it will join basis column Index
```

```python
def profit_flag(a) :
    if a > 0 :
        return 'positive'
    else :
        return 'negative'
```

```python
profit_flag(45)
```

```python
#Map,FIlter and Reduce function can be applied on lists but we want to apply such function for Series
df_sales['flag_profit'] = df_sales['profit'].apply(profit_flag) #Applying our function to each row of Dataframe
df_sales.head()
```

```python
df_sales['len_cust_name'] = df_sales['customer_name'].apply(len) #Applying inbuilt function
df_sales
```

```python
def quanity_flag(a):
    if a<10:
        return 'low'
    elif a>10 and a<20:
        return 'medium'
    else:
        return 'high'

df_sales['flag_quantity'] = df_sales['quantity'].apply(quanity_flag)
df_sales
```

```python
df_sales['square_quantity'] = df_sales['quantity'].apply(lambda a : a**2)
df_sales
```
